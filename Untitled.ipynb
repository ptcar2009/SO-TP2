```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./SO-TP2/test/output.csv")
```

```python
data.head(10)
```

```python
data['absolute'] = data.apply(lambda x: x['board_size']/x['page_size'], axis=1)
```

```python
data.head()
```

```python
for file in data['file'].unique():
    for alg in data['algorithm'].unique():
        queryed = data.query('algorithm == @alg and file == @file')[['exec_time', 'absolute']]
        exec_time = queryed['exec_time'].tolist()
        absolute = queryed['absolute'].tolist()

        ax = sb.lineplot(x = absolute, y = exec_time, label = alg, ci = None)
        ax.set_title('\n Arquivo de entrada: ' + file)
        ax.set(xlabel='Tamanho da tabela (# páginas)', ylabel='Tempo total (s)')
    plt.yticks(np.arange(min(absolute)-0.2, max(absolute)+0.2), 0.1)
    plt.show()
```

```python
for file in data['file'].unique():
    for alg in data['algorithm'].unique():
        queryed = data.query('algorithm == @alg and file == @file')[['page_faults', 'absolute']]
        page_faults = queryed['page_faults'].tolist()
        absolute = queryed['absolute'].tolist()

        ax = sb.lineplot(x = absolute, y = page_faults, label = alg, ci = None)
        ax.set_title('\n Arquivo de entrada: ' + file)
        ax.set(xlabel='Tamanho da tabela (# páginas)', ylabel='Page faults')
        
#     plt.xticks(np.arange(min(table_size), max(table_size)+1, 998))
    plt.show()
```

```python
for file in data['file'].unique():
    for alg in data['algorithm'].unique():
        queryed = data.query('algorithm == @alg and file == @file')[['dirty', 'absolute']]
        dirty = queryed['dirty'].tolist()
        absolute = queryed['absolute'].tolist()

        ax = sb.lineplot(x = absolute, y = dirty, label = alg, ci = None)
        ax.set_title('\n Arquivo de entrada: ' + file)
        ax.set(xlabel='Tamanho da tabela (# páginas)', ylabel='dirty count')
        
#     plt.xticks(np.arange(min(table_size), max(table_size)+1, 998))
    plt.show()
```
